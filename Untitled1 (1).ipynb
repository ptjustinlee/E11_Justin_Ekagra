```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inside_original = pd.read_csv('data.csv')
outside_modified = pd.read_csv('outside_full_sample.csv')
inside_new = pd.read_csv('inside_new_sample.csv')
outside_new = pd.read_csv('outside_new_sample.csv')

min_len = min(
    len(inside_original),
    len(outside_modified),
    len(inside_new),
    len(outside_new),
)

inside_original = inside_original.iloc[:min_len].reset_index(drop=True)
outside_modified = outside_modified.iloc[:min_len].reset_index(drop=True)
inside_new = inside_new.iloc[:min_len].reset_index(drop=True)
outside_new = outside_new.iloc[:min_len].reset_index(drop=True)

measurement_cols = inside_original.select_dtypes(include=np.number).columns

x = np.arange(min_len)

for col in measurement_cols:
    plt.figure()

    plt.plot(x, inside_original[col], label='Inside Original')
    plt.plot(x, inside_new[col], label='Inside New')
    plt.plot(x, outside_modified[col], label='Outside Modified')
    plt.plot(x, outside_new[col], label='Outside New')

    plt.xlabel("Measurement Index")
    plt.ylabel(col)
    plt.title(f"{col}: All Locations Comparison")
    plt.legend()
    plt.tight_layout()
    plt.show()

for col in measurement_cols:
    plt.figure()

    plt.hist(inside_original[col], bins=30, alpha=0.4, label='Inside Original')
    plt.hist(inside_new[col], bins=30, alpha=0.4, label='Inside New')
    plt.hist(outside_modified[col], bins=30, alpha=0.4, label='Outside Modified')
    plt.hist(outside_new[col], bins=30, alpha=0.4, label='Outside New')

    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"{col}: Distribution Comparison")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
